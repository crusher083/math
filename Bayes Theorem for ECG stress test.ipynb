**P(A)** - probability of angina pectoris in male in population in Poland = 0.097 ([source](https://www.ahajournals.org/doi/full/10.1161/CIRCULATIONAHA.107.720953#:~:text=Angina%20prevalence%20varied%20widely%20across,0.80%2C%20P%3C0.0001))<br>
**P(B|A)** - probability after positive ECG stress test - **SENSITIVITY**. The best estimation for ECG stress test=0.96 ([source](https://www.escardio.org/Education/Practice-Tools/EACVI-toolboxes/3D-Echo/accuracy-of-stress-echo))<br>
**P(B)** - probability of observing positive ECG stress test.

**P(B)** - is the sum of all positive test results. (True Positives(TP) and False Positives(FP)). 

**TP** = Sensitivity * Prevalence<br>
**FP** = (1 - Specificity) * (1 - Prevalence)<br>
Best estimation of **SPECIFICITY** if ECG stress test = 0.91 - ([source](https://www.escardio.org/Education/Practice-Tools/EACVI-toolboxes/3D-Echo/accuracy-of-stress-echo))

This way we will get the **PPV** formula: <br>

$$P(A|B) = {{prevalence \times sensitivity} \over {prevalence \times sensitivity + (1 - prevalence) \times (1 - specificity)}}.$$

Let's plug our numbers:

$$P(A|B) = {(0.097 \times 0.96) \over 0.097 \times 0.96 + (1 - 0.097) \times (1 - 0.92)}.$$

```python
p = (0.097 * 0.96) / ((0.097 * 0.96) + ((1 - 0.92) * (1 - 0.097)))
```

```python
print(f'The probability of angina pectoris in polish male with positive ECG stress test = {round(100*p,2)}%')
```
